## What is Scatter Plot?

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
%matplotlib inline
```

```python
sns.set()
```

```python
iris=sns.load_dataset("iris")
```

```python
iris.head()
```

```python
fig=plt.figure()
ax=plt.axes()
```

## Scatter Plot with Plot Method

```python
plt.plot(iris.sepal_length,iris.sepal_width, "o")
```

```python
plt.plot(iris.sepal_length,iris.sepal_width, "+")
```

```python
plt.plot(iris.sepal_length,iris.sepal_width, "v")
```

```python
plt.plot(iris.sepal_length,iris.sepal_width, "rv")
```

```python
plt.plot(iris.sepal_length,iris.sepal_width, 
         "rv", markersize=10)
```

## Scatter Plot with Scatter Method

```python
plt.scatter(iris.sepal_length,iris.sepal_width)
```

```python
plt.scatter(iris.sepal_length,
            iris.sepal_width,s=75)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=75, c="m")
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=75, c="m", marker="v")
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=75, c="m", 
            edgecolor="black", 
            linewidth=1)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width,
            s=75, c="m", 
            edgecolor="black", 
            linewidth=1, 
            alpha=0.5)
```

```python
color=np.random.randint(0,10,150)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=100, 
            c=color, 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=100, 
            c=color, 
            cmap="Reds", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=100, 
            c=color, 
            cmap="viridis", 
            edgecolor="black", 
            linewidth=1, 
            alpha=1)
cbar=plt.colorbar()
cbar.set_label("Levels")
```

```python
sizes=np.random.randint(50,200,150)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
plt.xlabel("sepal_width")
plt.ylabel("petal_length")
```

```python
plt.scatter(iris.sepal_length,iris.sepal_width, 
            s=sizes,  
            c=color, 
            cmap="Reds",
            edgecolor="black",
            linewidth=1, 
            alpha=1)
plt.xlabel("sepal_width")
plt.ylabel("petal_length")
plt.title("ƒ∞ris Sepal Length & Sepal Width ")
```
